```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
##2. Fit a sequence of AdaBoostClassifier with varying number of weak learners ranging 
##from 1 to 16, keeping the max_depth as 1. Plot the accuracy on test set against the 
##number of weak learners. Use decision tree classifier as the base classifier.
```

```python
from sklearn.tree import DecisionTreeClassifier
obj=DecisionTreeClassifier(criterion='entropy',max_depth=1)
from sklearn.ensemble import AdaBoostClassifier
boost=AdaBoostClassifier(base_estimator=obj,learning_rate=1)
boost.fit(x_train,y_train)
from sklearn.model_selection import GridSearchCV
neighbors={'n_estimators':list(range(1,17))}
gs=GridSearchCV(estimator=boost,param_grid=neighbors,scoring='accuracy',return_train_score=True)
gs.fit(x_train,y_train)
sca_x=gs.param_grid['n_estimators']
sca_x
```

```python
fd=pd.DataFrame(gs.cv_results_)
sca_y=np.array(fd['mean_test_score'])
import matplotlib.pyplot as plt
plt.scatter(sca_x,sca_y)
plt.show()
```
